### READING DATA

```python
df2021=pd.read_csv('universities V1.1.csv')
df2021.head()
```

```python
df2021['quality_of_education']=df2021['quality_of_education'].replace('-',0.0).astype(float)
df2021['alumni_employment']=df2021['alumni_employment'].replace('-',0.0).astype(float)
df2021['quality_of_faculty']=df2021['quality_of_faculty'].replace('-',0.0).astype(float)
df2021['research_performance']=df2021['research_performance'].replace('-',0.0).astype(float)
```
